```python
"""High level description of code

This script will use the csv library to read the budget data csv file, 
iterate over each row of the file to capture the date and PnL data, 
and write the below metrics to a text file.

The total number of months included in the dataset.
The net total amount of Profit/Losses over the entire period.
The average of the changes in Profit/Losses over the entire period.
The greatest increase in profits (date and amount) over the entire period.
The greatest decrease in losses (date and amount) over the entire period.

"""

# Import the pathlib and csv library
from pathlib import Path
import csv

# Initialize variable to hold pnl deltas
pnl_array = []
date_array = []

# Initialize metric variables
max_delta = 0
min_delta = 0
avg_delta = 0
total_pnl = 0
count_date = 0
pnl_delta = 0
pnl_delta_sum = 0
max_delta_date = 0
min_delta_date = 0

# Set the file path
csvpath = Path('./budget_data.csv')

#load csv and read
with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    header = next(csvreader)
    line_num += 1
    
#iterate rows with a for loop, ensure profit/losses are ints 
    for row in csvreader:
        pnl_variable = int(row[1])
        date_variable = str(row[0])
        pnl_array.append(pnl_variable)
        date_array.append(date_variable)
        
        
#calcalute sum of pnl, calculate total number of months
        count_date +=1
        total_pnl += pnl_variable
    
#iterate rows and calculate pnl deltas, store pnl deltas in a sum variable
        if count_date > 0:
            pnl_delta = pnl_array[count_date - 1] - pnl_array[count_date - 2]
            pnl_delta_sum += pnl_delta
            
            if pnl_delta > max_delta:
                max_delta = pnl_delta
                max_delta_date = date_array[count_date - 1]
                
            if pnl_delta < min_delta:
                min_delta = pnl_delta
                min_delta_date = date_array[count_date - 1]
                
#calculate average

avg_delta = pnl_delta_sum / (count_date - 1)
avg_delta = round(avg_delta,2)

#print answers
print(f"""
Financial Analysis
------------------------
""")
print(f"Total Months = {count_date}")
print(f"Total Profit and Loss = ${total_pnl}")
print(f"Average Change = ${avg_delta}")
print(f"Largest Daily Increase was ${max_delta}, which occured in {max_delta_date}")
print(f"Largest Daily Decrease was ${min_delta}, which occured in {min_delta_date}")

#print(pnl_delta) 
#print(pnl_delta_sum)
    
```
